```python
!pip install fuzzywuzzy
from fuzzywuzzy import process

# Assuming equity and list are your DataFrames
# Replace 'name' with the actual column names you want to search in both DataFrames

def find_similar_values(equity, stock_list, equity_name_column, stock_list_column):
    # Create a dictionary to store the matched values
    matched_values = {}

    # Iterate through each value in the stock_list DataFrame
    for stock in stock_list[stock_list_column]:
        if not isinstance(stock, str):
            break
        # Use process.extractOne to find the most similar value in the equity DataFrame
        match, score, index = process.extractOne(stock, equity[equity_name_column])
        # If the similarity score is above a certain threshold, consider it a match
        if score >= 70:  # You can adjust the threshold as needed
            matched_values[stock] = {'equity_name': match, 'equity_code': equity.loc[index, 'Security Code']}

    # Create a new DataFrame with the matched values
    result_df = pd.DataFrame(list(matched_values.items()), columns=[stock_list_column, 'matched_values'])
    result_df[['equity_name', 'equity_code']] = pd.DataFrame(result_df['matched_values'].tolist())
    result_df = result_df.drop(columns=['matched_values'])
    result_df['Asset'] = Stock_list['Asset']
    print(result_df)
    return result_df

# # Call the function with your DataFrames and the column names
# result_dataframe = find_similar_values(equity, stock_list, 'name', 'stock')
# print(result_dataframe)
```

```python
# Assuming you already have equity and stock_list DataFrames

# Call the function with your DataFrames and column names
result_dataframe = find_similar_values(Equity, Stock_list, 'Security Name', 'Name')

# Display the result DataFrame
print(result_dataframe)
```

```python
import time
# from BSE_module import BSE  # Assuming BSE_module is the module containing the BSE class

max_retries = 50
b = BSE()

def bsc(code, asset):
    try:
        quote = b.getQuote(str(code))
        data = []
        data.append(quote["companyName"])
        data.append(float(quote["previousClose"]))
        data.append(float(quote["currentValue"]))
        data.append(quote["updatedOn"])
        data.append(float(asset[:-1]))
        total_portfolio_value = 114.77
        stock_value = (float(asset[:-1]) / 100) * total_portfolio_value

        # Calculate change in stock value
        change_in_stock_value = (data[2] - data[1]) * (stock_value / data[1])

        # Calculate new portfolio value
        new_portfolio_value = total_portfolio_value + change_in_stock_value

        # Calculate percent change
        percent_change = (change_in_stock_value / total_portfolio_value) * 100
        data.append(percent_change)
        return data
    except Exception as e:
        print(f"Error processing code {code}: {e}")
        return None

quote = [["CompanyName","previousClose","currentValue","updatedOn","Asset","percent_change"]]
codelist = result_dataframe['equity_code']
assetlist = result_dataframe['Asset']
attempt = 0

for code, asset in zip(codelist, assetlist):
    while attempt < max_retries:
        try:
            bsc1 = bsc(code, asset)
            if bsc1 is not None:
                quote.append(bsc1)
                break  # Break the retry loop if successful
            else:
                print("Retrying...")
                attempt += 1
                time.sleep(1)  # Adding a delay between retries
        except Exception as e:
            print(f"Attempt {attempt + 1} failed with error: {e}")
            if attempt < max_retries - 1:
                print("Retrying...")
                attempt += 1
                time.sleep(1)  # Adding a delay between retries
            else:
                print("Max attempts reached. Exiting...")
                raise
```
